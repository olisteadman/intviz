#### violin

```python
def violin(
    wide=15, high=5,
    x='day',y='total_bill', palette="muted", split=False, set_style="whitegrid", bw='scott',
    scale='area', inner='box', linewidth=1, saturation=1
):
    try:
        fig, ax = plt.subplots()
        fig.set_size_inches(wide, high)
        sns.set_style(set_style)
        sns.violinplot(
        x=x, y=y, hue="smoker",
        data=tips,
        palette=palette, split=split, bw=bw, scale=scale, inner=inner, linewidth=linewidth,
        saturation=saturation
    );
    except Exception as e:
        print(e,"\33[1mPlease make sure you include at least 1 numeric value!")

interact(violin,
         linewidth=SelectionSlider(options=[("%g"%i,i) for i in reversed([i*10**-1 for i in range(21)])]),
         saturation=SelectionSlider(options=[("%g"%i,i) for i in reversed([i*10**-1 for i in range(11)])]),
         wide=widgets.IntSlider(min=5,max=15,step=1,value=10),
         high=widgets.IntSlider(min=5,max=15,step=1,value=5),
         split={bool:bool for bool in [True,False]},
         x={col:col for col in tips.columns},
         y={col:col for col in tips.columns},
         palette={col:col for col in [#'deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind',
                                      'Paired','BuGn','GnBu','OrRd','PuBu','YlGn','YlGnBu','YlOrBr','YlOrRd',
                                      'BrBG','PiYG','PRGn','PuOr','RdBu','RdGy','RdYlBu','RdYlGn','Spectral']},
         set_style={st:st for st in ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']},
         bw={col:col for col in ['scott', 'silverman']},
         scale={col:col for col in ['area', 'count', 'width']},
         inner={col:col for col in ['box', 'quartile', 'point', 'stick', None]}
         );
```

#### Complete Graph

```python
def graph(wide=15, high=5, n=10, with_labels=False, full_info=True):
    G = nx.complete_graph(n)
    plt.figure(figsize=(wide, high))
    
    if full_info:
        print("Edge count = triangular function (n²+n)/2 =\033[1m",len(G.edges()),"\033[0m",'\n')
        print("Adjacency matrix:")
        print(str(nx.adjacency_matrix(G).todense()).replace('0','\033[1m0\033[0m'),'\n')
        print("Edge list:\n",G.edges(), "\n")
        print("Node degrees:")
        print("From the adjacency matrix", nx.adjacency_matrix(G).sum(axis=0))
        print("Using networkx", G.degree(),'\n')
        print("Clustering coefficients:\n", nx.clustering(G))
        print("Average clustering coefficient", nx.average_clustering(G))
        print("Shortest path has length", nx.shortest_path_length(G, 1, 3), '\n',
              dict(nx.shortest_path_length(G)), '\n')
        node_centrality = nx.betweenness_centrality(G)
        edge_centrality = nx.edge_betweenness_centrality(G)
#         print('node_centrality:',node_centrality)
#         print('edge_centrality:',edge_centrality)
        
        plt.subplot(2,4,5)
        plt.barh(list(range(len(list(node_centrality.values())))),list(node_centrality.values()))
        plt.yticks(list(range(len(node_centrality))),list(node_centrality.keys()))
        plt.title('Node betweenness centrality');
        
        plt.subplot(2,4,6)
        plt.hist(list(node_centrality.values()))
        plt.title('Distribution of\nnode betweenness centrality');
        
        plt.subplot(2,4,7)
        plt.barh(list(range(len(list(edge_centrality.values())))),list(edge_centrality.values()))
        plt.yticks(list(range(len(edge_centrality))),list(edge_centrality.keys()))
        plt.title('Edge betweenness centrality');
        
        plt.subplot(2,4,8)
        plt.hist(list(edge_centrality.values()))
        plt.title('Distribution of\nedge betweenness centrality');
        
    plt.subplot(2,1,1)
    nx.draw_circular(G, with_labels=with_labels);

interact(graph,
         wide=widgets.IntSlider(min=5, max=15, step=1, value=15),
         high=widgets.IntSlider(min=5, max=15, step=1, value=10),
         n=widgets.IntSlider(min=1, max=50, step=1, value=10),
         );
```

#### Directed Graph

```python
def graph(wide=15, high=5, style='ggplot', n=10, rounds_of_edges=1, with_arrows=True, with_labels=False, full_info=False):
    G = nx.DiGraph()

    # define the nodes
    for i in range(n):
        G.add_node(i)

    # link the nodes
    for round in range(rounds_of_edges):
        for u, v in zip(random.sample(range(n), n), random.sample(range(n), n)):
            G.add_edge(u, v)

    max_edges = int((G.number_of_nodes()**2+G.number_of_nodes())/2)

    if full_info:
        print("Edge count =", G.number_of_edges(),
              "out of max trifunc (n²+n)/2 =\033[1m", max_edges, "\033[0m", '\n')
        print("Adjacency matrix:")
        print(str(nx.adjacency_matrix(G).todense()).replace(
            '0', '\033[1m0\033[0m'), '\n')
        print("Edge list:\n", G.edges(), "\n")
        print("Node degrees:")
        print("From the adjacency matrix ", nx.adjacency_matrix(G).sum(axis=0))
        print("Using networkx", G.degree(), '\n')
        print("In-degree: ", G.in_degree())
        print("Out-degree: ", G.out_degree())
        
        node_centrality = nx.betweenness_centrality(G)
        edge_centrality = nx.edge_betweenness_centrality(G)
#         print('node_centrality:',node_centrality)
#         print('edge_centrality:',edge_centrality)
        
        plt.subplot(2,4,5)
        plt.barh(list(range(len(list(node_centrality.values())))),list(node_centrality.values()))
        plt.yticks(list(range(len(node_centrality))),list(node_centrality.keys()))
        plt.title('Node betweenness centrality');
        
        plt.subplot(2,4,6)
        plt.hist(list(node_centrality.values()))
        plt.title('Distribution of\nnode betweenness centrality');
        
        plt.subplot(2,4,7)
        plt.barh(list(range(len(list(edge_centrality.values())))),list(edge_centrality.values()))
        plt.yticks(list(range(len(edge_centrality))),list(edge_centrality.keys()))
        plt.title('Edge betweenness centrality');
        
        plt.subplot(2,4,8)
        plt.hist(list(edge_centrality.values()))
        plt.title('Distribution of\nedge betweenness centrality');
        
    plt.style.use(style)
    plt.figure(figsize=(wide, high));

    plt.subplot(1, 3, 1)
    nx.draw_circular(G, arrows=with_arrows, with_labels=with_labels)

    plt.subplot(1, 3, 2)
    plt.hist(list(dict(G.degree()).values()),
             bins=np.linspace(0,
                              max(dict(G.degree()).values()),
                              max(dict(G.degree()).values())+1),
             align='left',rwidth=0.5)
    plt.xlabel('Degree')
    plt.ylabel('Counts')
    plt.title('Degree distribution');

    plt.subplot(1, 3, 3)
    plt.hist([list(dict(G.in_degree()).values()),list(dict(G.out_degree()).values())],
             bins=np.linspace(0,
                              max(dict(G.degree()).values()),
                              max(dict(G.degree()).values())+1),
             align='left',rwidth=0.5,stacked=False)
    plt.xlabel('Degree')
    plt.ylabel('Counts')
    plt.legend(['In-degree', 'Out-degree'],
               prop={'size': 10}, loc='upper left')
    plt.title('Degree distribution');

interact(graph,
         wide=widgets.IntSlider(min=5, max=15, step=1, value=15),
         high=widgets.IntSlider(min=5, max=15, step=1, value=5),
         n=widgets.IntSlider(min=1, max=50, step=1, value=10),
         rounds_of_edges=widgets.IntSlider(min=0, max=50, step=1, value=1),
         style={st: st for st in ['seaborn-dark', 'seaborn-darkgrid', 'seaborn-ticks', 'fivethirtyeight', 'seaborn-whitegrid', 'classic', '_classic_test', 'fast', 'seaborn-talk', 'seaborn-dark-palette', 'seaborn-bright', 'seaborn-pastel',
                                  'grayscale', 'seaborn-notebook', 'ggplot', 'seaborn-colorblind', 'seaborn-muted', 'seaborn', 'Solarize_Light2', 'seaborn-paper', 'bmh', 'tableau-colorblind10', 'seaborn-white', 'dark_background', 'seaborn-poster', 'seaborn-deep']},
         );
```

#### Weighted Graph

```python
from matplotlib import cm

def graph(wide=15, high=5, style='ggplot', n=10, rounds_of_edges=1, with_labels=False, full_info=False):
    G = nx.Graph()

    # define the nodes
    for i in range(n):
        G.add_node(i)

    # link the nodes
    for round in range(rounds_of_edges):
        for u, v in zip(random.sample(range(n), n), random.sample(range(n), n)):
            G.add_edge(u, v, weight=randint(0,10)/10)

    max_edges = int((G.number_of_nodes()**2+G.number_of_nodes())/2)

    if full_info:
        print("Edge count =", G.number_of_edges(),
              "out of max trifunc (n²+n)/2 =\033[1m", max_edges, "\033[0m", '\n')
        print("Adjacency matrix:")
        print(str(nx.adjacency_matrix(G).todense()).replace(
            '0', '\033[1m0\033[0m'), '\n')
        print("Edge list:\n", G.edges(), "\n")
        print("Node degrees:")
        print("From the adjacency matrix ", nx.adjacency_matrix(G).sum(axis=0))
        print("Using networkx", G.degree())

    plt.style.use(style)
    plt.figure(figsize=(wide, high));

    plt.subplot(1, 3, 1)
    nx.draw_circular(G, with_labels=with_labels)

    plt.subplot(1, 3, 2)
    plt.hist(list(dict(G.degree()).values()),
             bins=np.linspace(0,
                              max(dict(G.degree()).values()),
                              max(dict(G.degree()).values())+1),
             align='left',rwidth=0.5)
    plt.xlabel('Degree')
    plt.ylabel('Counts')
    plt.title('Degree distribution');

    plt.subplot(1, 3, 3)
    positions = nx.circular_layout(G)
    cmap_edges = cm.get_cmap('rainbow')
    edge_weights = [e[2]['weight'] for e in G.edges(data=True)]

    nx.draw(G,pos=positions,node_color='grey',with_labels=with_labels,
            edge_color=edge_weights,edge_cmap=cmap_edges)

    # code for adding a colour bar, colour range normalised to [0,1]
    vmin = 0
    vmax = 1
    sm = plt.cm.ScalarMappable(cmap=cmap_edges, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm._A = []
    plt.colorbar(sm);

interact(graph,
         wide=widgets.IntSlider(min=5, max=15, step=1, value=10),
         high=widgets.IntSlider(min=5, max=15, step=1, value=5),
         n=widgets.IntSlider(min=1, max=50, step=1, value=10),
         rounds_of_edges=widgets.IntSlider(min=0, max=50, step=1, value=1),
         style={st: st for st in ['seaborn-dark', 'seaborn-darkgrid', 'seaborn-ticks', 'fivethirtyeight', 'seaborn-whitegrid', 'classic', '_classic_test', 'fast', 'seaborn-talk', 'seaborn-dark-palette', 'seaborn-bright', 'seaborn-pastel',
                                  'grayscale', 'seaborn-notebook', 'ggplot', 'seaborn-colorblind', 'seaborn-muted', 'seaborn', 'Solarize_Light2', 'seaborn-paper', 'bmh', 'tableau-colorblind10', 'seaborn-white', 'dark_background', 'seaborn-poster', 'seaborn-deep']},
         );
```

#### Bipartite Graph

```python
from networkx.algorithms import bipartite

def graph(wide=15, high=5, n=10, rounds_of_edges=1, with_labels=False, full_info=False):
    B = nx.Graph()
    
    # add the node attribute "bipartite"
    B.add_nodes_from([i for i in range(int(n/2))], bipartite=0) 
#     print([i for i in range(int(n/2))])
    B.add_nodes_from([i for i in range(int(n/2),n)], bipartite=1)
#     print([i for i in range(int(n/2),n)])

    # link the nodes
    for round in range(rounds_of_edges):
        randlist = [(u,v) for u,v in list(zip([np.random.choice(range(int(n/2))) for n_ in range(n*2)],
                                              [np.random.choice(range(int(n/2),n)) for n_ in range(n*2)]))]
#         print(randlist)
        B.add_edges_from(randlist)
    
    bottom_nodes, top_nodes = bipartite.sets(B)
#     print(bipartite.sets(B))

    position = list(zip([0]*len(bottom_nodes),list(range(len(bottom_nodes)))))+list(zip([1]*len(top_nodes),list(range(len(top_nodes)))))
    print(position)

    positions = {}
    for node in bottom_nodes:
        positions[node] = np.array(position[node])
    for node in top_nodes:
        positions[node] = np.array(position[node])
        
    plt.figure(figsize=(wide, high));

    plt.subplot(1,4,1)
    nx.draw(B,pos=positions,with_labels=with_labels,node_color = ['red']*len(bottom_nodes) +
            ['blue']*len(top_nodes))
    
    G_top = bipartite.projected_graph(B, top_nodes)
    G_bottom = bipartite.projected_graph(B, bottom_nodes)
    
    plt.subplot(1,4,2)
    nx.draw_circular(G_bottom,with_labels=with_labels);
    
    plt.subplot(1,4,3)
    nx.draw_circular(G_top,with_labels=with_labels,node_color='blue');

    if full_info:
        print("First group:", bottom_nodes)
        print("Second group:", top_nodes)
        print(G_top.edges())
        print(G_bottom.edges())
        print("Adjacency matrix of the bipartite graph:") 
        print(nx.adjacency_matrix(B).todense())
        print("Adjacency matrix of the bipartite graph projected on group 0:") 
        print(nx.adjacency_matrix(B).dot(nx.adjacency_matrix(B)).todense()[:6,:6])
        print("Adjacency matrix of the bipartite graph projected on group 1:") 
        print(nx.adjacency_matrix(B).dot(nx.adjacency_matrix(B)).todense()[6:,6:])
        print("Degree list:\n", nx.degree(B))
        print("Degree list bottom nodes:",nx.degree(G_bottom))
        print("Degree list top nodes:",nx.degree(G_top))

    plt.subplot(1,4,4)
    
    # Generating the test graph with two components
    # First component
    nx.draw(B,with_labels=with_labels);

interact(graph,
         wide=widgets.IntSlider(min=5, max=15, step=1, value=10),
         high=widgets.IntSlider(min=5, max=15, step=1, value=5),
         n=widgets.IntSlider(min=1, max=50, step=1, value=10),
         rounds_of_edges=widgets.IntSlider(min=1, max=50, step=1, value=1),
         );
```

## Yellowbrick

```python
iris = sns.load_dataset("iris")
X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['species']
```

```python
# Can we make this interactive too? Same problem as above, need to let user select multiple inputs using checkboxes

from yellowbrick.features import Rank2D

visualizer = Rank2D(features=['sepal_length','sepal_width','petal_length','petal_width'], algorithm='covariance')
visualizer.fit(X, y)                # Fit the data to the visualizer
visualizer.transform(X)             # Transform the data
visualizer.poof()                   # Draw/show/poof the data
```

```python
from sklearn.svm import LinearSVC
from yellowbrick.classifier import ROCAUC

model = LinearSVC()
model.fit(X,y)
visualizer = ROCAUC(model)
visualizer.score(X,y)
visualizer.poof()
```
